```python
fig = plt.figure(figsize=(15,5))

sns.boxplot(x="loc2", y="biomass", data=df, hue='cultivar',linewidth=1)
plt.ylabel('Biomass ($t\,ha^{-1}$)', fontsize =20)
plt.xlabel('', fontsize =20)
ax = plt.gca()
ax.grid(True, 'major', color='w', linestyle='-', linewidth=0.5)
ax.grid(True, 'minor', color='0.92', linestyle='-', linewidth=0.7)
ax.patch.set_facecolor('0.96')
ax.set_axisbelow(True)
ax.yaxis.set_tick_params(labelsize=20)
ax.xaxis.set_tick_params(labelsize=13)
ax.set_xticklabels(ax.get_xticklabels(),rotation=0)
    
plt.legend(prop=dict(size=20),bbox_to_anchor=(1,1), loc=2, borderaxespad=0., frameon=False)
plt.subplots_adjust(wspace=0.2, hspace=0.2)
#plt.savefig(r'C:\Users\uqjojeda\Nextcloud\PURTERRA-A0131\2020\Validation\data.png', dpi=300,bbox_inches='tight')
```

```python
tkt_plot_detailed=sns.catplot(x="year", y="biomass", data=df, hue='cultivar', linewidth=1, kind='box', col='loc2', 
            col_wrap=4, legend=False)
tkt_plot_detailed.set_axis_labels("", "Biomass ($t\,ha^{-1}$)",fontsize= 25)
tkt_plot_detailed.set_titles(col_template="{col_name}", fontweight='bold', size=25)
for ax in tkt_plot_detailed.axes.flatten():

    ax.grid(True, 'major', color='w', linestyle='-', linewidth=0.5)
    ax.grid(True, 'minor', color='0.92', linestyle='-', linewidth=0.7)
    ax.patch.set_facecolor('0.90')
    ax.set_axisbelow(True)
    ax.yaxis.set_tick_params(labelsize=20)
    ax.xaxis.set_tick_params(labelsize=25)
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

plt.legend(prop=dict(size=20),bbox_to_anchor=(1.1, 0.8), loc=2, borderaxespad=0., frameon=False)
plt.subplots_adjust(wspace=0.03, hspace=0.15)

#plt.savefig(r'C:\Users\uqjojeda\Nextcloud\PURTERRA-A0131\2020\Validation\data2.png', dpi=300,bbox_inches='tight')
```
